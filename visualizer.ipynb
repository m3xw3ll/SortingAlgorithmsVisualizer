```python
# set array length
length = int(input('Please choose how many numbers you want to sort (50 is a good start):'))
speed = int(input('Please choose an animation speed (10 is a good start):'))
```

```python
while True:
    algorithm = int(input('Please choose an algorithm from above (1-12):'))
    if algorithm in range(1,12):
        break
    else:
        continue
```

```python
def update_fig(a, rects, iteration):
    """
     Function to update the plot
     :param a: the array with numbers
     :param rects: barrect for each number in the array
     :param iteration: the current iteration
     :return: None
     """
    for rect, val in zip(rects, a):
        rect.set_height(val)
    iteration[0] += 1
    text.set_text(f'# of operations {iteration}')
```

```python
# NOTE: This cell may take some time to finish depending which algorithm you choose and the array length

# create array
a = list(range(0,length))
# shuffle array
random.shuffle(a)
print(f'Your shuffled array is \n {a}')

# create generator
if algorithm == 1:
    generator = bubble_sort(a)
    name = 'Bubble Sort'
elif algorithm == 2:
    generator = cocktail_sort(a, 0, len(a) - 1)
    name = 'Cocktail Shaker Sort'
elif algorithm == 3:
    generator = comb_sort(a)
    name = 'Comb Sort'
elif algorithm == 4:
    generator = count_sort(a)
    name = 'Count Sort'
elif algorithm == 5:
    generator = gnome_sort(a)
    name = 'Gnome Sort'
elif algorithm == 6:
    generator = heap_sort(a)
    name = 'Heap Sort'
elif algorithm == 7:
    generator = insertion_sort(a)
    name = 'Insertion Sort'
elif algorithm == 8:
    generator = merge_sort(a, 0, len(a) - 1)
    name = 'Merge Sort'
elif algorithm == 9:
    generator = odd_even_sort(a)
    name = 'Odd Even Sort'
elif algorithm == 10:
    generator = quick_sort(a, 0, len(a) - 1)
    name = 'Quick Sort'
elif algorithm == 11:
    generator = selection_sort(a)
    name = 'Selection Sort'
elif algorithm == 12:
    generator = shell_sort(a)
    name = 'Shell Sort'
else:
    pass

fig, ax = plt.subplots()
ax.set_title(name)
bar_rects = ax.bar(range(len(a)), a, align='edge')
ax.set_xlim(0, length)
ax.set_ylim(0, int(1.07 * length))

text = ax.text(0.02, 0.95, "", transform=ax.transAxes)
iteration = [0]
animate = animation.FuncAnimation(fig,
                                  func=update_fig,
                                  fargs=(bar_rects, iteration),
                                  frames=generator,
                                  interval=speed,
                                  repeat=False,
                                  save_count=1500)
plt.close()
HTML(animate.to_html5_video())
```
